# scipy

```python
# Numercial arrays.
import numpy as np

# Scientific Computing.
import scipy as sp

# Plotting.
import matplotlib.pyplot as plt
```

## Fit a Line

```python
# Range of x values.
x = np.linspace(-10.0, 10.0, 21)

# Range of y values.
y = 4.0 * x + 2.0 + np.random.normal(0.0, 3.0, x.shape)
```

```python
# Plot x and y.
plt.plot(x, y, 'o', label='Data')
```

```python
# Fit a line.
m, c = np.polyfit(x, y, 1)

# Show.
m, c
```

```python
# Plot x and y.
plt.plot(x, y, 'o', label='Data')

# Plot the line.
plt.plot(x, m * x + c, label='Fit')
```

```python
# Fit a line.
coeffs = np.polyfit(x, y, 1)

# Show.
coeffs
```

```python
# Evaluate the polynomial.
f = np.polyval(coeffs, x)

# Show.
f
```

```python
# Show y.
y
```

## Evaluate

https://en.wikipedia.org/wiki/Coefficient_of_determination

```python
# Sum of squares of residuals.
ss_res = np.sum((y - f)**2)

# Show.
ss_res
```

```python
# Mean of y.
y_bar = np.mean(y)

# Show.
y_bar
```

```python
# Total sum of squares.
ss_tot = np.sum((y - y_bar)**2)

# Show.
ss_tot
```

```python
# Ratio.
ss_res / ss_tot
```

```python
# R^2 value.
1.0 - (ss_res / ss_tot)
```

## Using scipy

```python
# Use scipy to fit a line.
fit = sp.stats.linregress(x, y)

# Show.
fit
```

```python
# Slope.
fit.slope
```

```python
# Intercept.
fit.intercept
```

```python
# R value.
fit.rvalue
```

```python
# R2 value.
fit.rvalue**2
```
