## Pairs of Features

```python
# https://scikit-learn.org/stable/datasets/toy_dataset.html
diabetes = datasets.load_diabetes()

# Show
diabetes.keys()
```

```python
# Feature names.
diabetes.feature_names
```

```python
# Data.
diabetes.data
```

```python
# BMI.
bmi = diabetes.data[:, 2]

# Show first ten.
bmi[:10]
```

```python
# Blood pressure.
bp = diabetes.data[:, 3]

# Show.
bp[:10]
```

## Relationships

```python
# Create a figure and an axis.
fig, ax = plt.subplots()

# Scatter plot.
ax.scatter(bmi, bp)

# Labels.
ax.set_xlabel('BMI')
ax.set_ylabel('Blood Pressure')
```

## Colours

```python
# Create a figure and an axis.
fig, ax = plt.subplots()

# Scatter plot.
# Change marker:
# https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers
ax.scatter(bmi, bp, marker='.')

# Labels.
ax.set_xlabel('BMI')
ax.set_ylabel('Blood Pressure')
```

```python
# Sex.
sex = diabetes.data[:, 1]

# Show first ten.
sex[:10]
```

```python
# Unique sex values.
np.unique(sex)
```

```python
# Create a figure and an axis.
fig, ax = plt.subplots()

# Scatter plot.
# Change colour.
ax.scatter(bmi, bp, c=sex)

# Labels.
ax.set_xlabel('BMI')
ax.set_ylabel('Blood Pressure')
```
