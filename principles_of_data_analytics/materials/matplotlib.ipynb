## Plotting Functions

```python
# X values.
x = np.arange(0.0, 10.0, 0.1)

# Show
x
```

```python
# X values.
x = np.linspace(0.0, 100.0, 100)

# Show.
x
```

```python
# X values.
x = np.linspace(0.0, 10.0, 101)

# Show.
x
```

```python
# X values.
x = np.linspace(0.0, 100.0, 1001)

# Show.
x
```

```python
# Y values.
y = x**2

# Show.
y
```

```python
# Plot x vs y.
plt.plot(x, y);
```

```python
# Plot x vs y.
plt.plot(x, y, label='y = x^2');

# Add a legend.
plt.legend()

# Add labels.
plt.xlabel('x')
plt.ylabel('x^2')

# Add a title.
plt.title('A Lovely Plot');
```

```python
# Plot x vs y.
plt.plot(x, y, "g:", label='y = x^2');

# Add a legend.
plt.legend()

# Add labels.
plt.xlabel('x')
plt.ylabel('x^2')

# Add a title.
plt.title('A Lovely Plot');
```

## Comparing Plots

```python
# Domain of x values.
x = np.linspace(0.0, 10.0, 101)

# x squared
y1 = x**2

# 2 to the power of x.
y2 = 2**x
```

```python
# Plot x vs y1.
plt.plot(x, y1, label='y = x**2')

# Plot x vs y2.
plt.plot(x, y2, label='y = 2**x')

# Add a legend.
plt.legend()

# Add labels.
plt.xlabel('x')
plt.ylabel('y')

# Add a title.
plt.title('Two lovely plots');

```

```python
# Plot x vs y1.
plt.plot(x, y1, label='$ x^2$')

# Plot x vs y2.
plt.plot(x, y2, label='$ 2^x$')

# Add a legend.
plt.legend()

# Add labels.
plt.xlabel('x')
plt.ylabel('y')

# Add a title.
plt.title('Two lovely plots');
```

```python
# Domain of x values.
x = np.linspace(0.0, 100.0, 1001)

# x squared.
y1 = x**2

# 2 to the power of x.
y2 = 2**x
```

```python
# Plot x vs y1.
plt.plot(x, y1, label='$x^2$')

# Plot x vs y2.
plt.plot(x, y2, label='$2^x$')

# Add a legend.
plt.legend()

# Add labels.
plt.xlabel('x')
plt.ylabel('y')

# Add a title.
plt.title('Two Lovely Plots');
```

## Integers

https://numpy.org/doc/stable/reference/generated/numpy.arange.html

```python
# Domain of x values.

x = np.arange(1, 101, 1)

# Show.
x
```

```python
# Domain of x values
x = np.linspace(1, 100, 101, dtype=int)

# Show.
x
```

```python
# The type.
x.dtype
```

```python
# Domain of x values.
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
x = np.linspace(0, 10, 11, dtype=int)

# Show.
x
```

```python
# The type.
x.dtype
```

```python
# x cubed plus 2 x squared plus 1.
y = x**3 + 2 * x**2 + 1
```

```python
# Show y.
y
```

```python
# Type of y.
y.dtype
```

```python
# Plot x vs y.
plt.plot(x, y, '.', label='$x^3 + 2x^2 + 1$')

# Add a legend.
plt.legend()

# Add labels.
plt.xlabel('x')
plt.ylabel('y')

# Add a title.
plt.title('Plotting ints.');
```
